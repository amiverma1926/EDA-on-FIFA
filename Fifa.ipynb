```python
#Importing all the required libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Create the Bar plot of fifa dataset
plt.figure(figsize=(6,6))
plt.bar(list(fifa['nationality'].value_counts()[0:6].keys()),
        list(fifa['nationality'].value_counts()[0:6]))
plt.show()
```

```python
# Extract 2 Column from dataset
player_salary=fifa[['short_name','wage_eur']]
print(player_salary)
```

```python
player_salary.head()
```

```python
#sorting the data w.r.t wage_eur
player_salary=player_salary.sort_values(by=['wage_eur'],ascending=False)
print(player_salary)
```

```python
player_salary.head()
```

```python
# create a bar plot w.r.t top 5 players 
plt.figure(figsize=(7,5))
plt.bar(list(player_salary['short_name'].value_counts()[0:5].keys()),list(player_salary['wage_eur'].value_counts()[0:5]),color=['blue','pink','red','green','orange'])
plt.show()
```
